#**Random Numbers**

```python
import numpy as np
import matplotlib.pyplot as plt

#Normaly distributed numbers
NumbN = np.random.randn(200)
#Uniform distributed numbers
Numb = np.random.rand(200)


plt.subplot(121)
plt.plot(NumbN,'-')
plt.axis('square')
plt.title('Normaly distributed numbers')
plt.axis([0,20,-5,5])
plt.grid()
plt.subplot(122)
plt.plot(Numb)
plt.axis('square')
plt.title('Uniform distributed numbers')
plt.axis([0,20,-5,5])
plt.grid()

plt.show()
```

#**Converting between radians and degrees**

```python
import numpy as np

deg = 720

#rad = deg*np.pi/180
#rad = rad%(2*np.pi)

rad = np.deg2rad(deg)%(2*np.pi)

print('%g degrees is %g radians.' %(deg,rad))
```

```python
rad = np.pi

#deg = (180*rad/np.pi)%360
deg = np.rad2deg(rad)
print('%g radians is %g degrees.' %(rad,deg))
```

#**The Pythagorean theorem**

```python
import numpy as np
import matplotlib.pyplot as plt

a = 4
b = 3

c = np.sqrt(a**2 + b**2)

plt.plot([0,a],[0,0],'k',linewidth=2)
plt.plot([0,0],[0,b],'k',linewidth=2)
plt.plot([0,a],[b,0],'k',linewidth=2)
plt.plot([.3,.3],[0,.3],'k',linewidth=2)
plt.plot([0,.3],[.3,.3],'k',linewidth=2)

plt.text(a/3,-.30,'a='+str(a),fontsize=15)
plt.text(.1,b/3,'b='+str(b),fontsize=15)
plt.text(a/2+.1,b/2,'c='+str(c),fontsize=15)

plt.axis('square')
axlim = np.max((a,b)) +.5
plt.axis([-.5,axlim,-.5,axlim])
plt.grid()
plt.show()
```

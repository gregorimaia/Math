#**Graphing parabolas**

```python
import numpy as np
import matplotlib.pyplot as plt

#parameters
a = 1 # 1/4y
h = 1
k = -2
n = 100

# x-axis points to evaluate the function
x = np.linspace(h-5,h+5,n)

# create the function
y = a*(x-h)**2 + k

plt.plot(x,y)
plt.plot(y,x)
plt.grid()
plt.show()
```

#**Creating contours from meshes**

```python
import numpy as np
import matplotlib.pyplot as plt

X,Y = np.meshgrid(range(0,10),range(0,15))

plt.subplot(121)
plt.pcolormesh(X,edgecolor='k',linewidth=.1)

plt.subplot(122)
plt.pcolormesh(Y,edgecolor='k',linewidth=.1)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2,2,100)
X,Y = np.meshgrid(x,x)
s = 1

gauss2d = np.exp( -(X**2+Y**2)/s )

plt.imshow(gauss2d)
plt.show()
```

#**Graphing circles**

```python
import numpy as np
import matplotlib.pyplot as plt

h = 2
k = -3
r = 3

axlim = r + np.max((abs(h),abs(k)))
x = np.linspace(-axlim,axlim,100)
X,Y = np.meshgrid(x,x)

Fxy = (X-h)**2 + (Y-k)**2 - r**2

plt.contour(X,Y,Fxy,0)
plt.plot([-axlim,axlim],[0,0],'k--')
plt.plot([0,0],[-axlim,axlim],'k--')
plt.plot(h,k,'go')
plt.gca().set_aspect('equal')
plt.grid()
plt.show()
```

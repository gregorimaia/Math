#**Sumation and products**

```python
import numpy as np
import matplotlib.pyplot as plt

lst = [1,3,4,1,6]
print(np.sum(lst))
print(np.cumsum(lst))
print(np.prod(lst))
print(np.cumprod(lst))
plt.plot(lst)
plt.plot(np.cumsum(lst))
plt.plot(np.cumprod(lst))

plt.legend(['list','acumulated sum','acumulated product'])
plt.show()
```

#**Differences (discrete derivative)**

```python
import numpy as np
x = [2,4,6,-5,3,4]
np.diff(x)
```

```python
import numpy as np
#derivada segunda
v = np.arange(0,11)
print(v)
print(np.diff(v,2))
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-2,2,100)
y = x**2
dy = np.diff(y)
dx = np.diff(x)

plt.plot(x,y,label='f')
plt.plot(x[:-1],dy/dx,label='df/dx')
#para plotar a função e sua derivada, deve subtrair 1 termo
# de x, pois a derivada acaba sendo 1 termo menor que a função de origem.


plt.axis([-2,2,-1,2])
plt.legend()
plt.grid()
plt.show()
```

#**Quadratic equation**

```python
import numpy as np
import scipy as sp #when we have complex number, sqrt of a negative number
a = 3
b = 7
c = 5

quadeqP = (-b + sp.sqrt(b**2 - 4*a*c)) / (2*a)
quadeqN = (-b - sp.sqrt(b**2 - 4*a*c)) / (2*a)

print(quadeqP,quadeqN)
```

```python
import numpy as np
import scipy as sp

def quadeq(a,b,c):

    out = sp.zeros(2)

    out[0] = (-b + sp.sqrt(b**2 - 4*a*c)) / (2*a)
    out[1] = (-b - sp.sqrt(b**2 - 4*a*c)) / (2*a)

    return out

quadeq(2,7,5)
```

#**Complex numbers: conjugate and multiplication**

```python
import numpy as np
import sympy as sym

z1 = np.complex(4,5)
z2 = np.complex(6,-2)

w = z1*z2
c1 = np.conj(z1)
c2 = np.conj(z2)
#Numero complexo multiplicado pelo seu conjugado resultará em:
#(a + bj) * (a - bj) = a**2 + b**2
R2 = z2*c2
R1 = z1*c1

print(R1)
print(R2)
```

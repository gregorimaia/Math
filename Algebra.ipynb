#**Sumation and products**

```python
import numpy as np
import matplotlib.pyplot as plt

lst = [1,3,4,1,6]
print(np.sum(lst))
print(np.cumsum(lst))
print(np.prod(lst))
print(np.cumprod(lst))
plt.plot(lst)
plt.plot(np.cumsum(lst))
plt.plot(np.cumprod(lst))

plt.legend(['list','acumulated sum','acumulated product'])
plt.show()
```

#**Quadratic equation**

```python
import numpy as np
import scipy as sp #when we have complex number, sqrt of a negative number
a = 3
b = 7
c = 5

quadeqP = (-b + sp.sqrt(b**2 - 4*a*c)) / (2*a)
quadeqN = (-b - sp.sqrt(b**2 - 4*a*c)) / (2*a)

print(quadeqP,quadeqN)
```

```python
import numpy as np
import scipy as sp

def quadeq(a,b,c):

    out = sp.zeros(2)

    out[0] = (-b + sp.sqrt(b**2 - 4*a*c)) / (2*a)
    out[1] = (-b - sp.sqrt(b**2 - 4*a*c)) / (2*a)

    return out

quadeq(2,7,5)
```

#**Complex numbers: conjugate and multiplication**

```python
import numpy as np
import sympy as sym

z1 = np.complex(4,5)
z2 = np.complex(6,-2)

w = z1*z2
c1 = np.conj(z1)
c2 = np.conj(z2)
#Numero complexo multiplicado pelo seu conjugado resultará em:
#(a + bj) * (a - bj) = a**2 + b**2
R2 = z2*c2
R1 = z1*c1

print(R1)
print(R2)
```

#**The unit circle**

```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt

# Cria um array com 100 pontos igualmente espaçados no intervalo de 0 a 2π
x = np.linspace(0, 2*np.pi, 100)

# Plota um gráfico dos valores de cos(x) em relação a sin(x) usando a cor preta ('k')
plt.plot(np.cos(x), np.sin(x), 'k')

# Plota uma linha horizontal pontilhada no eixo x, indo de -1.1 a 1.1, com y igual a 0, usando a cor cinza claro ([.5, .5, .5])
plt.plot([-1.1, 1.1], [0, 0], '--', color=[.5, .5, .5])

# Plota uma linha vertical pontilhada no eixo y, indo de -1.1 a 1.1, com x igual a 0, usando a cor cinza claro ([.5, .5, .5])
plt.plot([0, 0], [-1.1, 1.1], '--', color=[.5, .5, .5])

# Define o ângulo como π/4 (45 graus)
angle = np.pi/4

# Plota uma linha vermelha ('r') a partir da origem até as coordenadas correspondentes ao cosseno e seno do ângulo definido
plt.plot([0, np.cos(angle)], [0, np.sin(angle)], 'r')

# Define a proporção dos eixos x e y como iguais, criando uma escala quadrada para o gráfico
plt.axis('square')

# Define os limites dos eixos x e y como -1.1 a 1.1, garantindo que o gráfico fique dentro dessa área
plt.axis([-1.1, 1.1, -1.1, 1.1])

# Define um rótulo para o eixo x como 'cos(x)'
plt.xlabel('cos(x)')

# Define um rótulo para o eixo y como 'sin(x)'
plt.ylabel('sin(x)')

# Exibe o gráfico resultante
plt.show()
```

#**Find a specific point on a Gaussian**

```python
import numpy as np
import matplotlib.pyplot as plt

h = .999

t = np.linspace(-3,3,1001)

#gaussian
g = np.exp( -4*np.log(2)*t**2 / h**2 )

#plot
plt.plot(t,g)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gaussian with FWHM=' + str(h)) #Full width at half maximum

plt.show()
yvalue = .5
tvalue = h * np.sqrt( np.log(yvalue) / (-4*np.log(2)) )
print(tvalue,-tvalue)
```

#**Log-spaced and linearly spaced numbers**

#**Arithmetic and geometric sequences.**

```python
import numpy as np
import matplotlib.pyplot as plt

# parameters
a = 10
d = 3
nmax = 10

# sequencia aritimetica
ariseq = a + d*np.arange(0,nmax)

# sequencia geometrica
b = 10
r = 3
geoseq = b * r**(np.arange(0,nmax))

print(ariseq)
print(geoseq)

plt.plot(ariseq,'ks', label='arithmatic')
plt.plot(geoseq,'ro', label='geometric')

plt.legend()
plt.show()
```

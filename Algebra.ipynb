#**Sumation and products**

```python
import numpy as np
import matplotlib.pyplot as plt

lst = [1,3,4,1,6]
print(np.sum(lst))
print(np.cumsum(lst))
print(np.prod(lst))
print(np.cumprod(lst))
plt.plot(lst)
plt.plot(np.cumsum(lst))
plt.plot(np.cumprod(lst))

plt.legend(['list','acumulated sum','acumulated product'])
plt.show()
```
